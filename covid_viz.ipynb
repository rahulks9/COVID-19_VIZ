## Graph-1

```python
X = df_new['Date_reported'].values
X.shape
```

```python
Y = df_new[' New_cases'].values
Y_tot = df_new[' Cumulative_cases'].values
Y.shape
```

```python
%matplotlib inline
#Xticks
xs = []
for i in range(156):
    if i%30==0:
        xs.append(X[i])
plt.style.use('fivethirtyeight')
plt.figure(figsize = (30,15))
plt.xlabel('Dates', fontsize =50)
plt.ylabel('Number of cases', fontsize =50)
plt.subplot(211)
plt.plot(X, Y)
plt.yscale('linear')
plt.title('linear scale', fontsize =50)
plt.grid(True)
plt.xticks(xs)
plt.xlabel('Dates')
plt.ylabel('Number of cases')
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize = (30,15))
plt.xlabel('Dates', fontsize =50)
plt.ylabel('Number of cases', fontsize =50)
plt.subplot(211)
plt.plot(X, Y_tot)
plt.yscale('linear')
plt.title('linear scale', fontsize =50)
plt.grid(True)
plt.xticks(xs)
plt.xlabel('Dates')
plt.ylabel('Number of cases')

plt.subplot(212)
plt.plot(X, Y_tot)
plt.yscale('log')
plt.title('logarithmic scale', fontsize =50)
plt.grid(True)
plt.xticks(xs)
plt.xlabel('Dates')
plt.ylabel('Number of cases')
plt.show()
```

# Graph6

```python
fp = "Admin2.shp"
map_df = gpd.read_file(fp)
map_df.head()
map_df.plot()
```

```python
map_df.head()
```

```python
merged = map_df.set_index('ST_NM').join(df_2.set_index('State'))
merged.head()
```

## Map 1

```python
variable = 'Total Cases'
vmin, vmax = merged['Total Cases'].min(), merged['Total Cases'].max()
fig, ax = plt.subplots(1, figsize=(10, 6))
merged.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
ax.axis('off')
ax.set_title('Total Cases Chloropleth')
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```

## Map2

```python
variable = 'Cured'
vmin, vmax = merged['Cured'].min(), merged['Cured'].max()
fig, ax = plt.subplots(1, figsize=(10, 6))
merged.plot(column=variable, cmap='Greens', linewidth=0.8, ax=ax, edgecolor='0.8')
ax.axis('off')
ax.set_title('Total Cured Chloropleth')
sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```

## Map3

```python
variable = 'Deaths'
vmin, vmax = merged['Deaths'].min(), merged['Deaths'].max()
fig, ax = plt.subplots(1, figsize=(10, 6))
merged.plot(column=variable, cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8')
ax.axis('off')
ax.set_title('Total Deaths Chloropleth')
sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```
