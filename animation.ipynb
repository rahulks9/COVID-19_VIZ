```python
list_lat= []
list_long = []

for i in range(3, 161):
    for k in range(data.shape[0]):
        if(int(data.iloc[k, i]) != 0):
            list_lat.append(data.iloc[k, 1])
            list_long.append(data.iloc[k, 2])
        else:
            continue
            
list_lat[0:5] 
list_long[0:5]
```

```python
len(list_lat)
```

```python
list_lat_uni = []
for x in list_lat:
    if x not in list_lat_uni:
        list_lat_uni.append(x) 
list_lat_uni
```

```python
list_long_uni = []
for y in list_long:
    if y not in list_long_uni:
        list_long_uni.append(y) 
list_long_uni[:-3]
```

## This makes a list of all the latitudes and longitudes, in the order of the COVID-19 timeline

## Animation

```python
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.io.shapereader as shpreader
import cartopy.io.img_tiles as cimgt
import matplotlib.pyplot as plt
import cartopy.mpl.geoaxes
import itertools
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
```

## Making the Cartopy Map

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
fig = plt.figure()
m4 = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
m4.add_feature(cfeature.LAND)
m4.add_feature(cfeature.OCEAN)
m4.add_feature(cfeature.COASTLINE)
m4.add_feature(cfeature.BORDERS, linestyle=':')
m4.add_feature(cfeature.LAKES, alpha=0.5)
m4.add_feature(cfeature.RIVERS)
```

## Making the animation and saving the file

```python
m4.stock_img()

X_points_bla = list_long_uni[:-3]
Y_points_bla= list_lat_uni

def animate(i):
    X_points= X_points_bla[0]
    Y_points= Y_points_bla[0]
    X_points_bla.remove(X_points_bla[0])
    Y_points_bla.remove(Y_points_bla[0])
    m4.scatter(X_points, Y_points, color = 'red')



ani= FuncAnimation(fig, animate, 256, interval=200, repeat = False)
ani.save('covid_ani.mp4', writer=writer)
plt.show()
```
