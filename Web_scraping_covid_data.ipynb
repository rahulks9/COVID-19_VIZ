```python
from bs4 import BeautifulSoup
import requests

with open('table.html') as html_file:
    soup = BeautifulSoup(html_file, 'lxml')
```

```python
State = []
Active_Cases = []
Cured = []
Deaths = []
Total_cases = []
for row in soup.find('tbody').find_all('tr'):
    cells = row.find_all('td')
    State.append(cells[1].text.rstrip('\n'))
    Active_Cases.append(cells[2].text.rstrip('\n'))
    Cured.append(cells[3].text.rstrip('\n'))
    Deaths.append(cells[4].text.rstrip('\n'))
    Total_cases.append(cells[5].text.rstrip('\n'))
```

```python
import pandas as pd
df = pd.DataFrame({'State' : State,
                  'Active Cases' : Active_Cases,
                  'Cured' : Cured,
                  'Deaths' : Deaths,
                  'Total Cases' : Total_cases})
df
```
